```python
actress_name_list = []
actress_link = []
actress_videos_list = []
actress_photos_list = []
details_6_bio = []
details_2_personal = []
details_3_rank = []
details_4_Aliases = []
details_5_website = []
for i in range(1,16):
    website = 'https://www.eporner.com/pornstar-list/'+str(i)+'/'
    html = urllib.request.urlopen(website).read()
    soup = BeautifulSoup(html,'lxml')
    actress = soup.find_all('div',class_='mbprofile')
    #print(len(actress))
    for j in actress:
        actress_name_list.append(j.a.text)
        actress_weblink = 'https://www.eporner.com'+str(j.a['href'])
        actress_link.append(actress_weblink)
        actress_videos_list.append(j.find('div',class_='mbtim').text.split()[-1])
        actress_photos_list.append(j.find('div',class_='mbvie').text.split()[-1])
        html = urllib.request.urlopen(actress_weblink).read()
        soup = BeautifulSoup(html,'lxml')
        #print(actress_weblink)
        details = soup.find('div',class_='psbio ps2')
        d1 = {}
        try:
            for k in details.ul.find_all('li'):
                d1[k.text.split(':')[0]] = k.text.split(':')[-1]
        except:
            d1 = np.NaN
        details_2_personal.append(d1)
        details = soup.find('div',class_='psbio ps3')
        d2 = {}
        try:
            for l in details.find_all('div'):
                d2[l.text.split(':')[0]] = l.text.split(':')[-1]
        except:
            d2 = np.NaN
        details_3_rank.append(d2)
        details = soup.find('div',class_='psbio ps4')
        d3 = []
        try:
            for m in details.div.ul.find_all('li'):
                d3.append(m.text)
        except:
            d3.append(np.NaN)
        details_4_Aliases.append(d3)
        details = soup.find('div',class_='psbio ps5')
        d4 = {}
        try:
            for n in(details.ul.find_all('li')):
                d4[n.text] = n.a['href']
        except:
            d4 = np.NaN
        details_5_website.append(d4)
        details = soup.find('div',class_='psbio ps6')
        try:
            d5 = details.div.text
        except:
            d5 = np.NaN
        details_6_bio.append(d5)
```
